We will be using a dataset from FiveThirtyEight's [Github repo](https://github.com/fivethirtyeight/data/tree/master/college-majors). It is based on the job outcomes of students who graduated college from 2010 to 2012.

We'd like to use visualizations to explore questions like:
1. Do students in more popular majors make more money? (scatter plot)
2. How many majors are predominantly male? Predominantly female? (histogram)
3. Which category of majors have the most students? (bar plot)

We'll use pandas and matplotlib.

### Setting Up The Environment

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

recent_grads = pd.read_csv("recent-grads.csv")

# Display the first row
recent_grads.iloc[0] 
```

### The columns explained:

Rank - Rank by median earnings (the dataset is ordered by this column).
Major_code - Major code.
Major - Major description.
Major_category - Category of major.
Total - Total number of people with major.
Sample_size - Sample size (unweighted) of full-time.
Men - Male graduates.
Women - Female graduates.
ShareWomen - Women as share of total.
Employed - Number employed.
Median - Median salary of full-time, year-round workers.
Low_wage_jobs - Number in low-wage service jobs.
Full_time - Number employed 35 hours or more.
Part_time - Number employed less than 35 hours.

### Pandas, Scatter Plots: Do students in more popular majors make more money?

Let's use a scatter plot to figure out if major popularity correlates with higher pay.

```python
recent_grads.plot(x='Sample_size', y='Median', kind='scatter', title='Sample_size vs. Median')
```

```python
recent_grads.plot(x='Sample_size', y='Unemployment_rate', kind='scatter', title='Sample_size vs. Unemployment_rate')
```

```python
recent_grads.plot(x='Full_time', y='Median', kind='scatter', title='Full_time vs. Median')
```

```python
recent_grads.plot(x='ShareWomen', y='Unemployment_rate', kind='scatter', title='ShareWomen vs. Unemployment_rate')
```

```python
recent_grads.plot(x='ShareWomen', y='Median', kind='scatter', title='ShareWomen vs. Median')
```

```python
recent_grads.plot(x='Men', y='Median', kind='scatter', title='Men vs. Median')
```

```python
recent_grads.plot(x='Women', y='Median', kind='scatter', title='Women vs. Median')
```

Based on the above charts, let's answer these questions line by line:
1. Do students in more popular majors make more money? The relationship does not seem strong. It appears that the smaller the sample size, the higher the median is.
2. Do students that majored in subjects that were majority female make more money? It appears that the more majority female a major is, it tends to have a lower median income.
3. Is there any link between the number of full-time employees and median salary? I don't think so because the median salary is based off of full-time employees.

### Pandas, Scatter Matrix Plot

Let's investigate the dataset using scatter matrix plots.

```python
from pandas.plotting import scatter_matrix 
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['Sample_size', 'Median', 'Unemployment_rate']], figsize=(10,10))
```

```python
scatter_matrix(recent_grads[['ShareWomen', 'Median']], figsize=(10,10))
```

With our charts side by side, we can revisit the questions:
1. Does major popularity correlate with higher median salary? No, it does not seem to be the case.
2. Do women-majority majors earn more? No, it appears that the median salary trends higher as there are less majority women.

### Pandas, Bar Plots

```python
recent_grads[:10].plot.bar(x='Major', y='ShareWomen', legend=False)
```

```python
recent_grads[-10:].plot.bar(x='Major', y='ShareWomen', legend=False)
```

According to bar charts of the first ten rows and last ten rows of this dataset (which is based on Median Salary), __we can conclude that there are less women in the top 10 paid majors. We can also conclude that the bottom 10 majors are majority women.__

```python
recent_grads[:10].plot.bar(x='Major', y='Unemployment_rate')
```

```python
recent_grads[-10:].plot.bar(x='Major', y='Unemployment_rate')
```

I don't see a clear trend between the top 10 vs. bottom 10 unemployment rates.

The top 10 has majority majors that do really well to be employed. Others like nuclear and mining go up to around 11-17% unemployment.

The bottom 10 is more spread out, hovering around an 7-8% average overall.

## Conclusions

From this project, we can conclude a couple of observations:
1. Popularity of a major does not correlate with a higher median salary.
2. The top 10 median salary majors tend to have a male majority.
3. The inverse is true: the bottom 10 median salary majors tend to have a female majority.
4. Unemployment is not solved completely whether or not your major is a top 10 or bottom 10 median salary major. There are still a lot of grads unable to find jobs in all majors.
